# Model Settings Sheep Win
 - Wolf gain from food = 20
 - show energy = False
 - initial number of wolves = 50
 - wolf reproduce = 5
 - intial number of sheep = 100
 - sheep reproduce = 4
 - model version = "sheep-wolves"
 - sheep gain from food = 4
 - grass regrowth time = 30

```python
import matplotlib.pyplot as plt
import pandas
```

```python
def sheepWin():
    FILENAME = "sheepwin.csv"
    tempData = pandas.read_csv(FILENAME)
    
    namesList=["Sheep Population","Wolf Population"]
    wolfY = list(tempData["wolfY"])
    sheepX = list(tempData["sheepX"])
    sheepY = list(tempData["sheepY"])
    
    plt.plot(sheepX,sheepY,label="Sheep Population") 
    plt.plot(sheepX,wolfY,label="Wolf Population")   
    plt.ylabel("Population")
    plt.xlabel("Time Ticks")
    plt.title("Population Over Time")
    plt.legend()
    plt.show()
sheepWin()
```

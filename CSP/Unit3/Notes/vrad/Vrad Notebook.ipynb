# Import file and necessary libraries
The file e are importing is a class that we created to reformat our vins. In a win there are 3 main sections...

 - wmi = first 3
 - vds = second 6
 - ser = third remaining items
 
 "vins.csv" is our original data

```python
from VIN import VIN

with open("vins.csv","r") as f:
    file = f.readlines()

for i in range(len(file)):
    file[i]=VIN(file[i].rstrip())
```

```python
#converts letters to base 10 number
def vmpenc(vin):
    vmap="0123456789ABCDEFGHJKLMNPRSTUVWXYZ"
    vrad=0
    exp=0
    base = 34
    print(vin)
    for v in range(len(vin)):
        coe = vmap.index(vin[v])
        exp = len(vin)-1-v
        vrad+= coe*base**exp
        #print(coe,exp,coe*base**exp)
    print(vrad)
    return(vrad)
```

```python
#Loop through your file and convert each item in the file
for v in file:
    vmpenc(v.wmi+v.vds+v.ser)
    
```
